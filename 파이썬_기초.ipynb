## 1.3 리스트와 리스트 연산
- 순서를 이용하여 값(데이터)을 정리하는 도구
- 대괄호[]를 이용, ','로 구분합니다.  
- 여러 종류의 값(데이터)이 혼합되어 올 수 있다.



````python
students = ["egoing", "sookbun", "duru"]
grades = [2, 1, 3]

print("students", students)
print("students[1]", students[1])
print("len(students)", len(students))
print("min(students)", min(students))
print("grades", grades)
print("min(grades)", min(grades))
print("max(grades)", max(grades))

# 리스트 연산: +, *
print("[1, 2, 3] + [4, 5, 6] =", [1, 2, 3] + [4, 5, 6])
print("[1, 2, 3] * 3 =", [1, 2, 3] * 3)
````

````python
# f(x) = 2x + 1
# f(3) => 7
# f(4) => 9
def f(x):
    return 2 * x + 1

print("f(3) =", f(3))
print("f(4) =", f(4))

def add(a, b):
    return a + b
print("add(3, 5) =", add(3, 5))

def message(name):
    return '안녕!! ' + name + ' ..., 환영합니다. ' + name + '!!'
print("message('숙번') =", message('숙번'))
print("message('이고잉') =", message('이고잉'))
````
